```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier

data = pd.read_csv("wine.csv", header = None)
                   #usecols=['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium']) 
                                        #'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
                                        #'Color intensity', 'Hue', 'OD280/OD315', 'Proline'])
data.head()
```

```python
target = data[0]
features = pd.DataFrame(data)
features = features.drop([0],axis=1)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
X = np.array(features)
y = np.array(target)
#neigh = KNeighborsClassifier(n_neighbors=1)
#neigh.fit(X,y)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale

X = scale(X)

for i in range(50):
    classifiers = []
# метод split возвращает индексы для объектов train и test
    for train_index, test_index in kf.split(y):
        neigh = KNeighborsClassifier(n_neighbors=i+1)
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        neigh.fit(X_train, y_train)
 
    # получим 5 оптимальных классификаторов
        classifiers.append(neigh)
    
        res = cross_val_score(neigh, X, y, cv=5)
 
    print(i+1,np.mean(res))
```
